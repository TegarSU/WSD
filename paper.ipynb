```python
# get ISEAR sentence for every ANEW

haduh2 = []
for index, row in anew.iterrows():
    def kerjo(haduh2):
        for indexs, rows in isear.iterrows():
            regex = r'(?<!\S)' + row['Description'] + r'(?!\S)'
            if ((str(row['Emotion']) == str(rows[36])) and bool(re.search(regex,rows.iloc[40].lower()))):
                row['Sentence'] = rows.iloc[40]
                haduh2.append(row)
                return
    kerjo(haduh2)

hasil2 = pandas.DataFrame(haduh2)
hasil2[['Description','Emotion','Sentence']].head()
#hasil2.head()
```

```python
# post tagging ANEW word in ISEAR sentences

def pos_tag(hasil2):

    tokens = nltk.word_tokenize(str(hasil2['Sentence']))
    pos_tags = nltk.pos_tag(tokens)
    # pos_tags
    pos_dict = {}
    for token, pos in pos_tags:
        pos_dict[token.lower()] = pos

    for key, value in pos_dict.items() :
        if(key == hasil2['Description']):
            return value
    return numpy.nan

hasil2['POS'] = hasil2.apply(lambda x: pos_tag(x), axis=1)
hasil2.head()
```

```python
hasil2[['Description','Emotion','Sentence','POS']].head()
```

```python
#define synset

hasil2=hasil2.dropna()
x=hasil2['POS'] # POS

#print (x)
def get_pos(x):
    if x.startswith('J'): #adj
        return 'a'
    elif x.startswith('V'): #verb
        return 'v'
    elif x.startswith('N'): #noun
        return 'n'
    elif x.startswith('R'):
        return 'r'
    else:
        return None
    #print (get_post(x))
    
def get_synset(hasil2): #get synset
    words = nltk.word_tokenize(hasil2['Sentence'])
    synset = lesk(words, hasil2['Description'], get_pos(hasil2['POS']))
    if synset:
        return synset.name()
    else:
        return 'NaN'

hasil2['Synset'] = hasil2.apply(lambda x:get_synset(x), axis=1)
hasil2.head()
```

```python
hasil2[['Description','Emotion','Sentence','POS','Synset']].head()
```

```python
y = hasil2['Sentence'] #ISEAR sentence for ANEW

def get_emotion(y):
    #y = hasil2['Sentence']
    words = nltk.word_tokenize(y) #tokenize ISEAR sentence
    #return words
    
    pos_tags = nltk.pos_tag(words) # POS for ISEAR sentence
    pos_dict = {}
    for token, pos in pos_tags:
        pos_dict[token] = pos
        
    val = 0
    ars = 0
    count = 0
    
    for word in words:
        word_df = anew.loc[anew['Description'] == word]
        if len(word_df) == 0:
            if get_pos(pos_dict[word]):
                word_df = anew.loc[anew['Description'] == WordNetLemmatizer().lemmatize(word, get_pos(pos_dict[word]))]
            
        if not word_df.empty:
            val += float(word_df['Valence Mean'])
            ars += float(word_df['Arousal Mean'])
            count += 1
    
    if count == 0:
        return 'NaN'
    
    val_avg = val/count
    ars_avg = ars/count
            
    if val_avg >= 5 and ars_avg >= 5:
        return 'joy'
    elif val_avg < 5 and ars_avg >= 5:
        return 'anger'
    elif val_avg < 5 and ars_avg < 5:
        return 'sadness'
    else:
        return 'other'  

def test_emotion(hasil2):
    synset = wordnet.synset(hasil2['Synset'])
    sentence = synset.definition()
    for example in synset.examples():
        sentence += '. ' + example
    return get_emotion(sentence)

test = hasil2[hasil2['Synset'] != 'NaN']
test = test[['Description', 'Synset', 'Emotion']].copy()
test['Emotion Test'] = test.apply(lambda x:test_emotion(x), axis=1)
print(test)
```
